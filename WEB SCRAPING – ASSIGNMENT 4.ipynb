1. Scrape the details of most viewed videos on YouTube from Wikipedia.

```python
driver=webdriver.Chrome(r"chromedriver.exe")
```

```python
driver.get("https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos")
```

```python
rank_no=[]
name_person=[]
artist_person=[]
upload_date=[]
views_num=[]
```

```python
try:
    rank=driver.find_elements(By.XPATH,'//table[@class="wikitable sortable jquery-tablesorter"][1]/tbody/tr/td[1]')
    for i in rank:
        rank_no.append(i.text)
except NoSuchElementException:
    rank_no.append('-')
try: 
    name=driver.find_elements(By.XPATH,'//table[@class="wikitable sortable jquery-tablesorter"][1]/tbody/tr/td[2]')
    for i in name:
        name_person.append(i.text)
except NoSuchElementException:
    name_person.append('-')
try:
    artist=driver.find_elements(By.XPATH,'//table[@class="wikitable sortable jquery-tablesorter"][1]/tbody/tr/td[3]')
    for i in artist:
        artist_person.append(i.text)
except NoSuchElementException:
    artist_person.append('-')
try:
    upload=driver.find_elements(By.XPATH,'//table[@class="wikitable sortable jquery-tablesorter"][1]/tbody/tr/td[5]')
    for i in upload:
        upload_date.append(i.text)
except NoSuchElementException:
    upload_date.append('-')
try:
    views=driver.find_elements(By.XPATH,'//table[@class="wikitable sortable jquery-tablesorter"][1]/tbody/tr/td[4]')
    for i in  views:
        views_num.append(i.text)
except NoSuchElementException:
    views_num.append('-')
```

```python
len(rank_no),len(name_person),len(artist_person),len(upload_date),len(views_num)
```

```python
import pandas as pd
df=pd.DataFrame({'Rank':rank_no,'Name':name_person,'Artist':artist_person,'Upload Date':upload_date,'Views':views_num})
df
```
